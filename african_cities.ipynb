The cities are still too many to work with, so let's reduce the number to around 200 by clustering. KMeans was used which is a clustering library by sklearn

```python
from sklearn.cluster import KMeans
```

```python
## the latitude and longitude of the cities will be the one used for the clustering

X = africa_cities[['lat', 'lng']]

# num_clusters defines the number of clusters to be created using the KMeans algorithm
num_clusters = 200

kmeans = KMeans(n_clusters=num_clusters) 

kmeans.fit(X)

#getting the labels for each cluster, the clusters are from 0-199
cluster_labels = kmeans.labels_

##adding the cluster column to the africa_cities dataframe
africa_cities['cluster'] = cluster_labels

```

```python
africa_cities.head()
```

```python
##Now let's select the most populated place in each of the cluster that will be used as our target area
def populated_target(clusters):
    max_population = clusters['population'].idxmax()
    return clusters.loc[max_population]
```

```python
target_points = []

for cluster in range(num_clusters):
    clusters = africa_cities[africa_cities['cluster'] == cluster]
    most_populated = populated_target(clusters)
    target_points.append(most_populated)
    
target_points = pd.DataFrame(target_points)
    
```

```python
target_points = target_points.reset_index(drop=True)
target_points.head(20)
```

```python
target_points.to_csv('targets.csv')
```
